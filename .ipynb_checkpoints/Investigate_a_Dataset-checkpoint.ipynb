## Exploratory Data Analysis



### 1. is there a certain trends that appear in the movies that generate higher profits.

```python
profit = (df.revenue - df.budget) > df.budget
print(df[profit]['revenue'].corr(df[profit]['popularity']))
```

```python
plt.figure(figsize=(8, 8), dpi=80)
plt.scatter(x=df[profit]['revenue'], y=df[profit]['popularity'])
plt.xlabel('REVENUE')
plt.ylabel('POPULARITY')
plt.title(' POPULARITY V. REVENUE', fontSize = 30)
```

###### revenue and popularity have a strong positive correlation. a movie that is popular is expected to generate more profit

```python
plt.figure(figsize=(30, 30), dpi=20)
plt.bar(x = df[profit]['genres'], height= df[profit]['revenue'])
plt.xlabel('REVENUE')
plt.ylabel('GENRES')
plt.title(' GENRES V. REVENUE', fontSize = 30)
```

###### some genres are expected to generate more profit when compared to others

### 2. does the voting score effect the revenue of a certain movie

```python
plt.scatter(y= df.revenue, x=df.vote_average)
print(df.revenue.corr(df.vote_average))
plt.ylabel('REVENUE')
plt.xlabel('VOTE AVERAGE')
plt.title(' REVENUE V. VOTES', fontSize = 30)
```

######  as we can see from the above graph voting doesn't really effect the revenue of a certain movie because they show very weak positive correlation

```python
plt.hist(df['revenue'])
plt.xlabel('revenue')
plt.ylabel('# movies')
df['revenue'].describe()
```
